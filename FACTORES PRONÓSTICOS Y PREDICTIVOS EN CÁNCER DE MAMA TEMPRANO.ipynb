```python
import matplotlib.pyplot as plt

edad = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]
casos_por_edad = [5, 12, 18, 25, 30, 35, 28, 22, 15, 10]

plt.figure(figsize=(8, 5))
plt.bar(edad, casos_por_edad, color='skyblue')
plt.title("Distribución de Casos por Edad")
plt.xlabel("Edad (años)")
plt.ylabel("Número de Casos")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
tamano_tumoral = ["<1cm", "1-2cm", "2-5cm", ">5cm"]
casos_tamano = [20, 35, 30, 15]

plt.figure(figsize=(8, 5))
plt.bar(tamano_tumoral, casos_tamano, color='lightcoral')
plt.title("Distribución por Tamaño Tumoral")
plt.xlabel("Tamaño Tumoral")
plt.ylabel("Número de Casos")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
cirugia_axilar = ["No", "Sí"]
casos_cirugia = [40, 60]

plt.figure(figsize=(6, 6))
plt.pie(casos_cirugia, labels=cirugia_axilar, autopct='%1.1f%%', colors=['gold', 'lightgreen'], startangle=90)
plt.title("Proporción de Pacientes con Cirugía Axilar")
plt.show()
```

```python
receptores_hormonales = ["Positivo", "Negativo"]
casos_receptores = [70, 30]

plt.figure(figsize=(6, 6))
plt.pie(casos_receptores, labels=receptores_hormonales, autopct='%1.1f%%', colors=['lightblue', 'pink'], startangle=90)
plt.title("Estado de Receptores Hormonales")
plt.show()
```
