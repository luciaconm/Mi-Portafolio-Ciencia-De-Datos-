```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
csv_path = "datos_combinados.csv"  # Asegúrate de que la ruta sea correcta
df = pd.read_csv(csv_path, header=None)

# Identificar la columna de rendimiento (última columna probable)
rendimiento_col = df.columns[-1]

# Convertir la columna a valores numéricos (por si hay valores no numéricos)
df[rendimiento_col] = pd.to_numeric(df[rendimiento_col], errors='coerce')

# Crear histograma
plt.figure(figsize=(10, 6))
plt.hist(df[rendimiento_col].dropna(), bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Rendimiento (KG/HA)')
plt.ylabel('Frecuencia')
plt.title('Histograma de Rendimiento de Soja')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar gráfico
plt.show()

# Imprimir mensaje de confirmación
print("El histograma de rendimiento de soja ha sido generado correctamente.")
```
