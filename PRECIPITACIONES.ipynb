```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
csv_path = "datos_combinados.csv"  # Asegrate de que la ruta sea correcta
df = pd.read_csv(csv_path, header=None)

# Identificar las columnas de tiempo y precipitaciones (suponiendo que la primera es el ao y las siguientes meses)
tiempo_col = df.columns[0]  # Primera columna como referencia temporal
precipitaciones_cols = df.columns[1:-1]  # Columnas intermedias como precipitaciones
rendimiento_col = df.columns[-1]  # ltima columna como rendimiento

# Convertir las columnas a valores numricos
df[precipitaciones_cols] = df[precipitaciones_cols].apply(pd.to_numeric, errors='coerce')
df[rendimiento_col] = pd.to_numeric(df[rendimiento_col], errors='coerce')

# Crear grfico de lneas de precipitaciones
df.set_index(tiempo_col)[precipitaciones_cols].plot(figsize=(12, 6))
plt.xlabel('Tiempo')
plt.ylabel('Precipitaciones (mm)')
plt.title('Precipitaciones a lo largo del tiempo')
plt.legend(title='Meses')
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar grfico
plt.show()
print("El grfico de lneas de precipitaciones ha sido generado correctamente.")

# Crear diagrama de dispersin entre precipitaciones y rendimiento
plt.figure(figsize=(10, 6))
plt.scatter(df[precipitaciones_cols].sum(axis=1), df[rendimiento_col], color='blue', alpha=0.5)
plt.xlabel('Total de Precipitaciones (mm)')
plt.ylabel('Rendimiento (KG/HA)')
plt.title('Diagrama de Dispersin: Precipitaciones vs Rendimiento')
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar grfico
plt.show()
print("El diagrama de dispersin entre precipitaciones y rendimiento ha sido generado correctamente.")
```
